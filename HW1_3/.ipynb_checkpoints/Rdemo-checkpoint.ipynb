## Q2. 折價是否影響買東西的金額

```r
result_pay <- tapply(order$UnitPrice , order$OrderGroupCode , sum)
result_dis <- tapply(order$PromotionDiscount , order$OrderGroupCode , sum)
result_pay <- data.frame(result_pay)
result_dis <- data.frame(result_dis)
result <- cbind(result_pay,result_dis)
result$dis <- (-result$result_dis)
```

```r
head(result)
```

```r
qplot(result$result_pay, result$dis,geom = c("point", "smooth"))
```

## Q3. 取消訂單最嚴重的縣市
#### 取消訂單與訂單數量在各縣市的比例

```r
s <- subset(order,StatusDef == 'Finish',select = 'City')
sn <- subset(order,StatusDef != 'Finish',select = 'City')
```

```r
tables <- as.data.frame(table(s))
tablesn <- as.data.frame(table(sn))
tables <- subset(tables,Freq!=0)
tablesn <- subset(tablesn,Freq!=0)
#dim(tables) == dim(tablesn)
names(tables)[names(tables)=="Freq"]="Finish"
names(tablesn)[names(tablesn)=="Freq"]="NotFinish"
#tablesn$s
#tables$s
citystatus <- merge(x = tables, y = tablesn, by.x = "s", by.y = "sn", all.x = TRUE)
citystatus$rate <- citystatus$NotFinish/citystatus$Finish
head(citystatus)
```

```r
options(warn=-1)
qplot(s, data = citystatus, geom = "bar", weight = rate) + ylab("rate")+ 
    theme(text=element_text(family="STHeitiTC-Light", size=6))
options(warn=1) 
```
